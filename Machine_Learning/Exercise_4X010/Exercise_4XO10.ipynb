```python
import pandas as pd
```

```python
file_path = "/Volumes/SAYAN/Cuvette-Data-Science/Machine_Learning/Exercise_4X010/car price.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
print(df)
```

```python
# Display the first few rows and info
df.head(), df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
pd.set_option('display.max_columns', None)
```

```python
# Copy the original dataframe
df_encoded = df.copy()

# 1. Label Encoding for binary column 'Leather_Interioe'
label_encoder = LabelEncoder()
df_encoded['Leather_Interioe'] = label_encoder.fit_transform(df_encoded['Leather_Interioe'])

# 2. One-Hot Encoding for nominal categorical columns
one_hot_columns = [
    'Manufacturer', 'Model', 'Category', 'Fuel_Type', 
    'Gear_Box_Type', 'Drive_Wheel', 'Wheel', 'Color'
]

df_encoded = pd.get_dummies(df_encoded, columns=one_hot_columns)

# Display the shape and first few rows of the encoded dataset
df_encoded.shape, df_encoded.head()
```

```python
# Save the encoded DataFrame to a new CSV file
encoded_file_path = "/Volumes/SAYAN/Cuvette-Data-Science/Machine_Learning/Exercise_4X010/car_price_encoded.csv"
df_encoded.to_csv(encoded_file_path, index=False)

encoded_file_path
```
